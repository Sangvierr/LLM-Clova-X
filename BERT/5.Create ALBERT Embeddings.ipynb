### 5.Create ALBERT Embeddings

````python
# 라이브러리
import torch
from transformers import AlbertTokenizer, AlbertModel
````

````python
# 모델 불러오기
model = AlbertModel.from_pretrained('albert-base-v2')
````

````python
# AlbertTokenizer requires the SentencePiece library
# pip install sentencepiece
````

````python
# 토크나이저 불러오기
tokenizer = AlbertTokenizer.from_pretrained('albert-base-v2')
````

- 모델과 토크나이저를 불러왔다.
- 이제 입력 문장에 대한 입력 값을 얻어보자.

````python
# 입력
sentence = "Paris is a beautiful city"
````

````python
# 토크나이저에 입력값 입력
inputs = tokenizer(sentence, return_tensors='pt')
````

````python
# 토큰화 결과 출력
print(inputs)
````

````python
# 모델에 토큰 값을 넣는다.
hidden_rep, cls_head = model(**inputs, return_dict = False)
````

````python
# 은닉 상태 크기 확인
print(hidden_rep.shape)
````

- [배치 크기, 시퀀스 길이, 은닉 상태 크기]이다.
- BERT 모델과 똑같다.

````python
# 결과 확인 - 첫번째 배치의 첫번째 단어의 임베딩 값
print(hidden_rep[0][0])
````

````python
# 결과 확인 - 첫번째 배치의 두번째 단어의 임베딩 값
print(hidden_rep[0][1])
````

또한 다음과 같이 각 토큰의 표현을 얻을 수 있다.
- ```hidden_reps[0][0]```은 첫번째 배치의 첫번째 시퀀스이므로 '[CLS]'의 임베딩 벡터이다.
- ```hidden_reps[0][1]```은 첫번째 배치의 두번째 시퀀스이므로 'Paris'의 임베딩 벡터이다.
- ```hidden_reps[0][6]```은 첫번째 배치의 마지막 시퀀스이므로 '[SEP]'의 임베딩 벡터이다.
